```
### read input
data_sample =  pd.read_csv("input.txt",sep=',',header = None)
data_sample = data_sample.values.tolist()
print("Input data:")
for i in range(0,5):
    print(data_sample[i])

### print second output(without first integer) #
data_output = pd.read_csv("output.txt",sep = '|', header = None)
data_output_flag = data_output[:][0].tolist()
data_output = data_output[:][1]
data_output_split = data_output.str.split(':',expand=True)
print()
print("Output data:")
print(data_output_flag[0:5])
```

```
### print type of first sequence(C or D) #

data_output_seqtype = data_output_split[0]
data_output_seqtype=data_output_seqtype.tolist()
print(data_output_seqtype[0:5])
```

```
### print the regression number
### 输出第一列回归数据
import re

data_seq1=data_output_split[1].tolist()
seq1=[]
p1 = re.compile(r'[(](.*?)[)]') #find out the content in ()#
for sequence in data_seq1:
    temp=re.findall(p1, sequence)
    temp1=''.join(temp)
    temp2=temp1.split(',')
    temp2=list(map(eval,temp2))
    seq1.append(temp2)

print("Processing data:")
for i in range(0,5):
    print(data_seq1[i])
print("\nAfter split:")
for i in range(0,5):
    print(seq1[i])
```

```
### Devide Train set and Output set

### Type: ONE C item
input_one = []
output_one = []
### Type C sequence
input_c = []
output_c = []
### Type D sequence
input_d = []
output_d = []
### Valid
split_buffer = []

for i in range (0,len(data_sample)):
    if data_output_flag[i] == 0:
        input_one.append(data_sample[i])
        output_one.append(seq1[i])
    elif data_output_flag[i] != 0:
        if data_output_seqtype[i] == "C":
            input_c.append(data_sample[i])
            output_c.append(seq1[i])
        elif data_output_seqtype[i] == "D":
            input_d.append(data_sample[i])
            output_d.append(seq1[i])
        else:
            split_buffer.append(seq1[i])
    else:
        split_buffer.append(seq1[i])

if not split_buffer:
    print("Divided successful.")
```

```
### result of one_set
if len(input_one) == len(output_one):
    print("Training Set:")
    for i in range(0,5):
        print(input_one[i])
    print("\nOutput Set:")
    for i in range(0,5):
        print(output_one[i])
    print("\nLength of one_set: "+str(len(input_one)))
else:
    print("Error length.")
```

```
### result of TypeC_set
if len(input_c) == len(output_c):
    print("Training Set:")
    for i in range(0,5):
        print(input_c[i])
    print("\nOutput Set:")
    for i in range(0,5):
        print(output_c[i])
    print("\nLength of TypeC_set: "+str(len(input_c)))
else:
    print("Error length.")
```

```
### result of TypeD_set
if len(input_d) == len(output_d):
    print("Training Set:")
    for i in range(0,5):
        print(input_d[i])
    print("\nOutput Set:")
    for i in range(0,5):
        print(output_d[i])
    print("\nLength of TypeD_set: "+str(len(input_d)))
else:
    print("Error length.")
```

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from sklearn.metrics import accuracy_score
```

```
class nn_classic(nn.Module):
    def __init__(self, input_size, output_size):
        super(nn_classic, self).__init__()
        self.linear1 = nn.Linear(input_size, 10)
        self.linear2 = nn.Linear(10, 7)
        self.linear3 = nn.Linear(7, 5)
        self.linear4 = nn.Linear(5, output_size)

    def forward(self, x2):
        x2 = self.linear1(x2)
        x2 = self.linear2(F.relu(x2))
        x2 = self.linear3(F.relu(x2))
        Zout = self.linear4(F.relu(x2))
        return Zout

model = nn_classic(13, 5)
learning_rate=0.001
no_of_epochs = 1000

criterion = nn.CrossEntropyLoss()
loss1 = nn.MSELoss() 
optimiser = optim.SGD(model.parameters(), lr=learning_rate)#===========================

for epoch in range(no_of_epochs):  # loop over the dataset multiple times
    
    for i in range(0, 1000):

        inputs_torch = torch.from_numpy(np.array(input_c[i])).float()
        output_torch = torch.from_numpy(np.array(output_c[i])).float()

        # get the inputs
        inputs = inputs_torch
        labels = output_torch

        model.train()
        # zero the parameter gradients
        optimiser.zero_grad()

        # forward + backward + optimize
        outputs = model(inputs)
        loss = loss1(outputs, labels) # We don't need to calcualte logsoftmax here
        loss.backward()
        optimiser.step()

    # print statistics
    if epoch % 20 == 19:    # print every 2 epochs
        model.eval()
        pred_outputs = model(inputs)
        #predicted = torch.argmax(pred_outputs, 1)
        #train_acc = accuracy_score(predicted.numpy(),np.array(nn_output_seq[i]))
        print('%d, loss: %.4f' %(epoch + 1, loss.item()))

#print('Finished Training')
```
